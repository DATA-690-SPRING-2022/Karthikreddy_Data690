```python
df_pop_age = df.query("indicator == 'SP.POP.1564.TO.ZS'")
df_pop_age = df_pop_age[df_pop_age["Country Code"].isin(["BFA","CAF", "ETH", "LBR","MDG","TGO","UGA"])]
i2 = df_pop_age["value"]
i5 = df_pop_age["Year"]

df_unemp_tot = df.query("indicator == 'SL.UEM.TOTL.ZS'")
df_unemp_tot = df_unemp_tot[df_unemp_tot["Country Code"].isin(["BFA","CAF", "ETH", "LBR","MDG","TGO","UGA"])]
i3 = df_unemp_tot["value"]

df2 = pd.DataFrame({ 'Year': i5.values, 'POP': i3.values, 'UNEMP': i2.values})
```

```python
fig = px.scatter(df2, x="UNEMP", y="POP",labels={"UNEMP":"Unemployment Total", "POP":"Population ages"},
    height=800,
    trendline="lowess",
    text="Year",
    title="Umemployment vs Total Population",
    template=list(plotly.io.templates.keys())[5])
fig.update_layout(showlegend=True)
fig.update_traces(textposition="bottom center")

fig.show()
```
